## 1. Background

Floating point (FP) is a system to represents real numbers and carrying out arithmetic operations in modern machine learning pipline. For example, pytorch has default precesion of FP32 on CPU. FP is mainly related to storage and precesion, one thing is often neglected is the summation/subtraction operation accumulate **error** in a multiplicative factor. A self contained reference by __[DAVID GOLDBERG](https://docs.oracle.com/cd/E19957-01/800-7895/800-7895.pdf)__. 
For the following content, we assume that readers have read through above contents. 

For a machine have $\beta = 2$ as base, $p$ digits of mantissa, $e$ as exponents, a real number $x$ that is $p+k$ digits FP representatble, we can write $x$ as  
$$ x = x_0. x_1 x_2 \cdots x_{p-1} x_{p} \cdots x_{p+k} \times \beta^{e}$$
the subscipts are just for counting the digits in this document.

We use $x-y$ to represent exact subtraction and $x \ominus y$ to represent FP subtraction. The operation of $x \ominus y$ behind the machine is:
1. reprents both $x$ and $y$ in FP using the same $e$:
$$ x = x_0. x_1 x_2 \cdots x_{p-1} x_{p} \cdots x_{p+k} \times \beta^{e} $$
$$ y = y_0. y_1 y_2 \cdots y_{p-1} y_{p} \cdots y_{p+k} \times \beta^{e}$$
2. truncate both FP to the $(p+1)^{th}$ digits:
$$ \tilde{x} = x_0. x_1 x_2 \cdots x_{p-1} x_{p} \times \beta^{e} $$
$$ \tilde{y} = y_0. y_1 y_2 \cdots y_{p-1} y_{p} \times \beta^{e}$$
3. perform subtraction using the truncated mantissa only:
$$ \tilde{x} - \tilde{y} = (x_0. x_1 x_2 \cdots x_{p-1} x_{p} - y_0. y_1 y_2 \cdots y_{p-1} y_{p}) \times \beta^{e} $$
4. round up/down dependens to the ${p}^{th}$ digits:
$$ x \ominus y = \lfloor \tilde{x} - \tilde{y} \rceil = \alpha_0. \alpha_1 \cdots \alpha_{p-1} \times \beta^{e} $$ 

## 2. Error

From above 4 steps illustration of FP subtraction, step 2 and step 4 can contributed to arithimetic **error** when training a machine learning model. It makes sense to measure the **error** using the notation of relative error:
$$ \epsilon = \frac{|(x \ominus y) - (x-y)|}{x-y} $$
since FP arithemic was operated on the mantissa only. According to __[DAVID GOLDBERG](https://docs.oracle.com/cd/E19957-01/800-7895/800-7895.pdf)__, if $x$ is exactly FP representable, $\epsilon  < \beta^{-p+1}$.

In this notebook, we observe the errors in FP operation within the context of machine learning. But first we will start with the context of FP subtraction only.

**case 1**: both $x$ and $y$ are FP representable, there is no error. Since no truncating error after $(p+1)^{th}$ digit in step 2, and no rounding error from $(p+1)^{th}$ digit to the $p^{th}$ digit. 

How can we know whether both $x$ and $y$ are both FP representable? We can test it by ratio. For $x$ and $y$ being both FP representable, we have 
$$\frac{x}{y} \leq \frac{\max x}{\min y} = \frac{1_0.1_1 \cdots 1_{p-1} \times \beta^{e}}{0_0. 0_1 \cdots 0_{p-2} 1_{p-1} \times \beta^{e}} = \frac{1 - \beta^{p+1}}{(1-\beta)}$$.

We will use floating point 16 where $p = 10$, (also recall for this document we assumned $\beta = 2$ by defalt, which is the base number used in all FP represntations), hence we only need to test whether $x/ y \leq 2047 $ 

    
    

```python
# parameter of FP:
p = int(10) 
beta = int(2)
ratio = (beta**(p+1)- int(1))/(beta-int(1))

print('the above threshold is computed as {}'.format(int(ratio)))
```

We then test some numerical values of $x$ and $y$ whose ratio us below this threshold, the result of $x \ominus y$ is as expected no error:

```python
x = torch.tensor(16, dtype = torch.float16)
y = torch.tensor(1, dtype = torch.float16)

print('the ratio of x/y is {} and x-y = {}'.format(x/y, x-y))
```

Does it mean if $x = 2048$ and $y=1$ then the error is guaranteed? The answer is no.

```python
x = torch.tensor(2048, dtype = torch.float16)
y = torch.tensor(1, dtype = torch.float16)

print('the ratio of x/y is {} and x-y = {}'.format(x/y, x-y))
```

Although the ratio exceeded the threshold, error was not observed. The reason is that both step 2 and 4 are exact coincidently due to $\beta = 2$:
1. $x = (2048)_{(10)} = (1.0000000000 \times 2^{10})_{(2)}, y = 1_{(10)} = (0.0000000001 \times 2^{10})_{(2)}$
2. note both $x, y$ has $(p+1)$ digits hence $\tilde{x}, \tilde{y}$ are same as above
3. $\tilde{x}-\tilde{y} = 0.1111111110 \times 2^{10}$ which is exact.
4. round to $p^{th}$ digit hence $x \ominus y = 0.111111111 \times 2^{10} = 2047$.

**case 2**: when $x$ is FP representable, $y$ is not representable using $(p+1)$ digits, error occurs.

Similary, we can catch this case by testing whether the ratio between $x$ and $y$ are above the threshold 2048.

```python
x = torch.tensor(2050, dtype = torch.float16)
y = torch.tensor(1, dtype = torch.float16)

print('the ratio of x/y is {} and x-y = {}'.format(x/y, x-y))
```

As illustrated above, the result is no longer as expected due to errors in step 2 or/and step 4. There is a special senario under this case, which is when $x$ and $y$ are different in magnitude **too much**, $x \ominus y = x$. When does this case occur? when ratio is greater than $2^{p+2} = 2^{12}$. Here is an example:

```python
x = torch.tensor(4096, dtype = torch.float16)
y = torch.tensor(1, dtype = torch.float16)

print('the ratio of x/y is {} and x-y = {}'.format(x/y, x-y))
```

The explanation for the above example is that in step 2, $x/y \leq \beta^{p+2}$, which implies that $\tilde{y} = 0 \times \beta^{e}$. Hence $x \ominus y = x$. This case is particularly interesting in the context of machine learning. Hence, some stopping creterion is built based on model performance/parameters across time. **When model parameters are no longer updating due to FP errors, whether terminating training procedure due to FP error implies that the model has been trained to the optimal in practice?**

## 3. Stochastic Gradient Descent with FP
In the context of model training with stochastic gradient descent (sgd), ideally we update model parameters with 
$$w_{t+1} = w_{t} - \eta_t g_t$$
where $\eta_t$ is an non-negative learning rate, $g_t = \nabla f(x_t)$ is the gradient of loss function $f(\cdot)$ evaluated at a training instance $x_t$. In reality, instead of exact subtraction we perform FP subtraction as illustrated in previous section:
$$w_{t+1} = w_{t} \ominus \eta_t g_t$$
This suggests that given we start the algorithm such that $w_0$ is FP representable, $w_t$ will always be FP representable.

Now, we will simulate a non-linear separable binary classification problem with logistic loss, and we assume $\eta_t = 1 $:

```python
class Model:
    # w: weights
    # n: dimension
    # dtype : '16' or '32'
    
    # x/y: data/label

    def __init__(self, n, dtype):
        
        if (dtype == 16):
            self.dtype = torch.float16
        else:
            self.dtype = torch.float32  
        self.w = torch.zeros(n, dtype = self.dtype)
        self.t = 1
        

    def fit(self, x, y):
        x = x.type(self.dtype)
        y = y.type(self.dtype)
        y_pred = torch.dot(self.w,x)
        
        l = self.loss(y, y_pred)
        dl = self.grad(x, y, y_pred)
        
        assert l.dtype == self.dtype
        assert dl.dtype == self.dtype
        
        self.w = self.w - (1/1) * dl
        assert self.w.dtype == self.dtype
        
        self.t += 1
        
        return self.w, l, dl
        
        
        
    def loss(self, y, y_pred):
        return np.log10(1 + np.exp(-y*y_pred))
    
    def grad(self, x, y, y_pred):
        return (-y * x)/(1+ np.exp(-y*y_pred)) 
        
```

The dataset used as shown below: 

```python
def half_moon(n_samples=1000, noise=0.1, random_state=0, to_plot = True):
        datasets = make_moons(n_samples=n_samples, noise=noise, random_state=random_state)

        X, y = datasets
        X = StandardScaler().fit_transform(X)
        X = X.astype(float)
        
        indx = y==0
        y[indx] = -1

        if to_plot:
            cm = plt.cm.RdBu
            cm_bright = ListedColormap(['#FF0000', '#0000FF'])

            fig, ax = plt.subplots()
            ax.set_title("Input data")

            ax.scatter(X[:, 0], X[:, 1], c=y, cmap=cm_bright,
                       edgecolors='k')

            #ax.set_xticks(())
            #ax.set_yticks(())

            plt.tight_layout()
            plt.show()
            fig.savefig('data.png', dpi = 300)
        return torch.Tensor(X), torch.Tensor(y)
data, label = half_moon()
```

We train the model for a fixed epoch number and record $w_t$ and $g_t$:

```python
w_hist = list()
l_hist = list()
dl_hist = list()

model = Model(2, 16)
for epoch in range(20):
    indx = np.arange(len(label))
    np.random.shuffle(indx)
    for t in indx:
        x = data[t]
        y = label[t]
        
        w,l, dl = model.fit(x,y)
        
        w_hist.append(w)
        l_hist.append(l)
        dl_hist.append(dl)

y_pred = (torch.matmul(w,data.type(torch.float16).t()))>=0
y = label == 1
print('Classification accuracy: {}%'.format((y_pred.type(torch.uint8) == y).sum()/1000*100))
```

We can check coordinate wise magnitude of $w_t, g_t$, here we show the first coordinate:

```python
w1 = [w[0] for w in w_hist]
w2 = [w[1] for w in w_hist]

dl1 = [dl[0] for dl in dl_hist]
dl2 = [dl[1] for dl in dl_hist]

#t = np.arange(20000)+1
#g1 = [dl[0]*1/t for dl, t in zip(dl_hist,t)]
fig,ax = plt.subplots()
ax.plot(w1,color = 'blue')
# set x-axis label
ax.set_xlabel("t")
# set y-axis label
ax.set_ylabel("w1", color = 'blue')

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(dl1, color = 'red', alpha=0.3)
ax2.set_ylabel("g1", color = 'red')
plt.show()
#fig.savefig('w_g_lr.png', dpi = 300)
```

The above plot shos that the magnitude difference between $w1$ and $g1$ gradually different from each other, it also seems that $w1$ does not update anymore after around $t = 5000$. In fact, it is true that $w1$ did not update. Let's plot it in a different way, we show both $w1$ and $w1/g1$ across time, which confirms that if $w1/g1 > 2^{11}$, $w_t \ominus g_t = w_t $, and the weight doesn't update anymore due to FP error in step 2.

```python
ratio =  [np.log2(np.abs(i / j)) for i, j in zip(w1, dl1)]
fig,ax = plt.subplots()
ax.plot(w1,color = 'blue')
# set x-axis label
ax.set_xlabel("t")
# set y-axis label
ax.set_ylabel("w1", color = 'blue')

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(ratio, color = 'red', alpha=0.3)
ax2.set_ylabel("Ratio : $2^k$", color = 'red')
ax2.hlines(11,0, 20000, color = 'black', linestyles='dashed')
plt.show()
#fig.savefig('ratio_lr.png', dpi = 300)
```

In practice, the stopping criterion often relies on training/testing accuracy. We use stored $w_t$ to check the accuracy at each step $t$. It turns out that we would probably terminated the training process before $t=5000$ since there is no significant change in model accuracy anyway. This is saying that terminating the training process due to FP error when $w_t$ is no longer updating is probably unlikely. Hence, **when the terminating criterion based on model performance is met, FP error is not likely to be a concern.**

```python
acc_hist = list()
for i in w_hist:
    w = i
    y_pred = (torch.matmul(w,data.type(torch.float16).t()))>=0
    y = label == 1 
    acc = (y_pred.type(torch.uint8) == y).sum()/1000
    
    acc_hist.append(acc)
    
fig,ax = plt.subplots()
ax.plot(acc_hist,color = 'blue')
# set x-axis label
ax.set_xlabel("t")
# set y-axis label
ax.set_ylabel("acc", color = 'blue')

ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(ratio, color = 'red', alpha=0.3)
ax2.set_ylabel("Ratio : $2^k$", color = 'red')
ax2.hlines(11,0, 20000, color = 'black', linestyles='dashed')
plt.show()
#fig.savefig('acc_lr.png', dpi = 300)
```
